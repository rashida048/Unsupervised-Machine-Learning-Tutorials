```python
X_values = df[['Size (cm)', 'Weight (g)']].values
```

```python
X_values
```

```python
from sklearn.cluster import KMeans 
```

```python
inert = []
for i in range(1, 19):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=32)
    kmeans.fit(X_values)
    inert.append(kmeans.inertia_)
```

```python
inert
```

```python
plt.plot(range(1,19), inert)

plt.xlabel('The No of clusters')

plt.ylabel('Outcome')

plt.title('The Elbow Method')

plt.show()
```

```python
kmeans = KMeans(n_clusters=2, init='k-means++')
kmeans.fit(X_values)

kmeans.cluster_centers_
```

```python
y_cl = kmeans.fit_predict(X_values)
y_cl 
```

```python
plt.scatter(X_values[y_cl == 0,0], X_values[y_cl == 0, 1], c = 'red', s = 100)

plt.scatter(X_values[y_cl == 1,0], X_values[y_cl == 1, 1], c = 'green', s = 100)

#plt.scatter(X_val[y_kmeans == 2,0], X_val[y_kmeans == 2, 1], c = 'orange', s = 100)

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300,
           c='brown')

plt.title('K-Means Unsupervised Learning')

plt.show()
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++')
kmeans.fit(X_values)

kmeans.cluster_centers_
```

```python
y_cl = kmeans.fit_predict(X_values)
y_cl 
```

```python
plt.scatter(X_values[y_cl == 0,0], X_values[y_cl == 0, 1], c = 'red', s = 100)

plt.scatter(X_values[y_cl == 1,0], X_values[y_cl == 1, 1], c = 'green', s = 100)

plt.scatter(X_values[y_cl == 2,0], X_values[y_cl == 2, 1], c = 'orange', s = 100)

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300,
           c='brown')

plt.title('K-Means Unsupervised Learning')

plt.show()
```
